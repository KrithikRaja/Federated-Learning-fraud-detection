```python
df = df_transaction_train.merge(df_identity_train, on='TransactionID', how='left')
df.drop(columns=['TransactionID'], inplace=True)
```

```python
df['card1'] = df['card1'].fillna(-1)
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.fillna("missing", inplace=True)
```

```python
df['BankID'] = df['card1'].astype(int) % 3
```

```python
categorical_features = df.select_dtypes(include=['object']).columns.tolist()
encoder = LabelEncoder()
for col in categorical_features:
    df[col] = encoder.fit_transform(df[col])
```

```python
numerical_features = [col for col in df.select_dtypes(include=[np.number]).columns if col != 'BankID']
scaler = StandardScaler()
df[numerical_features] = df[numerical_features].astype(float)
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python
df['card1'] = df['card1'].astype(int)
df['BankID'] = df['card1'] % 3
```

```python
for i in range(3):
    bank_df = df[df['BankID'] == i].drop(columns=['BankID'])
    
    if not bank_df.empty:
        bank_df.to_csv(f"preprocessed_bank_{i}.csv", index=False)
        print(f"Saved preprocessed_bank_{i}.csv with {len(bank_df)} rows.")
    else:
        print(f"Bank {i} dataset is empty and was not saved.")

print("Preprocessing complete.")
```
